```python
import pandas as pd
import psycopg2
import sqlalchemy
from sqlalchemy import create_engine
import yaml
from yaml import Loader

class DatabaseConnector:
    def __init__(self):
        self.yaml_file = open('db_creds.yaml', 'r')
        self.data = yaml.load(self.yaml_file, Loader=Loader)

    def read_db_creds(self):
        self.yaml_file
        self.data
        return self.data

    def init_db_engine(self):
        read_db = DatabaseConnector.read_db_creds(self)
        return create_engine(f"postgresql+psycopg2://{read_db['RDS_USER']}:{read_db['RDS_PASSWORD']}@{read_db['RDS_HOST']}:{read_db['RDS_PORT']}/{read_db['RDS_DATABASE']}")

    def list_db_tables(self):
        data = DatabaseConnector.read_db_creds(self)
        with psycopg2.connect(host=data['RDS_HOST'], user=data['RDS_USER'], password=data['RDS_PASSWORD'], dbname=data['RDS_DATABASE'], port=data['RDS_PORT']) as conn:
            with conn.cursor() as cur:
                cur.execute('''SELECT table_name FROM information_schema.tables WHERE table_schema = 'public';''')
                for table in cur.fetchall():
                    print(table)

    def read_db_table(self):
        data = DatabaseConnector.read_db_creds(self)
        con = psycopg2.connect(user = data['RDS_USER'], password = data['RDS_PASSWORD'], host = data['RDS_HOST'], port = data['RDS_PORT'], database = data['RDS_DATABASE'])
        engine = DatabaseConnector.init_db_engine(self)
        sql = 'SELECT * FROM legacy_users LIMIT 5;'
        df = pd.read_sql_query(sql,con)
        return df

    

final = DatabaseConnector()
final.read_db_table()
```
