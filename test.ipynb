```python
from sqlalchemy import create_engine
import pandas as pd
import psycopg2
import yaml
from yaml import Loader
from database_utils import DatabaseConnector

# Extract data from multiple sources
class DataExtractor:
    def __init__(self):
        pass

    # Read a Table
    def read_db_table(self, engine, table_name):
        df = pd.read_sql_table(table_name, engine)
        return df

    # Clean a Table
    def clean_user_data(self, engine, table_name):
        clean_table = DataExtractor.read_db_table(self, engine, table_name)
        clean_table.set_index("index", inplace=True)
        clean_table['address'] = clean_table['address'].str.replace('\n|/|\.|-|,', ' ', regex=True)
        clean_table['phone_number'] = clean_table['phone_number'].str.replace('\.|\(0\)|\(|\)|x', ' ')
        clean_table.dropna()
        return clean_table.head(60)

a = DataExtractor()
a.clean_user_data(engine, 'legacy_users')
```
